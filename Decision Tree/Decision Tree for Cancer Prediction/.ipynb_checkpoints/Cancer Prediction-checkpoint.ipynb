# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import pandas as pd
import numpy as np

#Wisoncsion Breat Cancer Dataset from UCI machine learning repository
data = pd.read_csv('data.csv')
data = data[['radius_mean','concave points_mean','diagnosis']]
data.head()
```

```python
#Vector of features
X = data[['radius_mean','concave points_mean']]

#We want the diagnosis data to be numerical: i.e. map M -> 1 and B -> 0
mapping = {'M':1,'B':0}
y = data['diagnosis'].map(mapping)

y.head()
```

```python
#Set a random seed for reproducible results
SEED=np.random.seed(343)
```

```python
#We now want to split our data into training set and test set
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size = .2, stratify = y, random_state = SEED)
```

```python
#Next we want to setup and train the model
dtc = DecisionTreeClassifier(max_depth = 6, random_state = SEED)
#fit the training data
dtc.fit(X_train,y_train)
#predict the test set labels
y_pred = dtc.predict(X_test)
print(y_pred[0:5])
```

```python
#We need a measurement of how "good" the model will preform
from sklearn.metrics import accuracy_score

y_pred = dtc.predict(X_test)

acc = accuracy_score(y_pred,y_test)
print('Test set accuracy: {:.2f}'.format(acc))
```

Not terrible, using only two features we were able to get a 84% accuracy on our model.

```python
#Since we do not have a UI we shall workinag with text
from sklearn.tree import export_text
text_representation = export_text(dtc)
print(text_representation)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (25,20))
_ = plot_tree(dtc, feature_names = ['Radius Mean','Concave Points Mean'], class_names = 'Diagnosis', filled = True)
fig.savefig('decision_tree.png')
```

# Entropy Criterion for DTC


```python
dtc_entropy = DecisionTreeClassifier(max_depth = 10, criterion = 'entropy' ,random_state = SEED)
dtc_entropy.fit(X_train,y_train)
y_pred_entropy = dtc_entropy.predict(X_test)
acc_entropy = accuracy_score(y_pred_entropy,y_test)
print('The accuracy score of DTC with entropy criterion is: {:.2f}'.format(acc_entropy))
```

# Gini Criterion for DTC

```python
dtc_gini = DecisionTreeClassifier(max_depth = 10, criterion = 'gini' ,random_state = SEED)
dtc_gini.fit(X_train,y_train)
y_pred_gini = dtc_gini.predict(X_test)
acc_gini = accuracy_score(y_pred_gini,y_test)
print('The accuracy score of DTC with gini criterion is: {:.2f}'.format(acc_gini))
```

# Quick LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state = SEED)

logreg.fit(X_train,y_train)

y_pred_logreg = logreg.predict(X_test)
acc_logreg = accuracy_score(y_pred_logreg,y_test)
print('The accuracy score of Logistic Regression is: {:.2f}'.format(acc_logreg))
```

# Ensemble Implementations

### Out of Bag (OOB)

```python
from sklearn.ensemble import BaggingClassifier

#Since the different models (w/ different criterion) performed about the same, we will pick any of them
#I chose the entropy model.

#Instantiate BaggingClassifier
bc = BaggingClassifier(base_estimator = dtc_entropy, n_estimators = 50, oob_score = True, random_state = SEED)

#Fit BaggingClassifier to the data
bc.fit(X_train,y_train)

#We already have the accuracy score of the dtc_entropy model
#To get the accuracy score of the OOB: extract it from bc
acc_oob = bc.oob_score_

#Print the results
print('Test set accuracy: {:.3f}, OOB accuracy: {:.3f}'.format(acc_entropy, acc_oob))
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, min_samples_leaf = .27, random_state = SEED)
rf.fit(X_train,y_train)
acc_rf = rf.score(X_test,y_test)
print('Random Forest Classifier Score {:.3f}'.format(acc_rf))
```

```python
#How do we know which feature provided the most predictive power?

#The following code creates a panda series object with the value of importance of each feature
importances = pd.Series(data=rf.feature_importances_, index= X_train.columns)
importances = importances.sort_values()

importances.plot(kind = 'barh', color = 'orange')
plt.title('Feature Importance')
plt.show()
                
```

### Adaptive boosting (AdaBoosting)

```python
# Import AdaBoostClassifier
from sklearn.ensemble import AdaBoostClassifier

#Instatiate ada
ada = AdaBoostClassifier(base_estimator=dtc_entropy, n_estimators=180, random_state=1)
ada.fit(X_train,y_train)
y_pred_proba = ada.predict_proba(X_test)[:,1]
```

```python
#Evaluatiing the AdaBoost classifier

#An ROC curve (receiver operating characteristic curve) is a graph showing the performance of a classification model at all classification thresholds. This curve plots two parameters:
    #True Positive Rate
    #False Positive Rate

# Import roc_auc_score
from sklearn.metrics import roc_auc_score

# Evaluate test-set roc_auc_score
ada_roc_auc = roc_auc_score(y_test, y_pred_proba)

# Print roc_auc_score
print('ROC AUC score: {:.2f}'.format(ada_roc_auc))
```

### AUC score WARNING!!!

In this example, it is important to note that ROC AUC score may not be desirable. Since false negatives are not weighed enough. What do I mean? Because we are dealing with cancerous tumors, we likely want to minimize false negatives as much as possible. otherwise it could lead to a misdiagnosis VERY BAD)
